```python
import pandas as pd
import wikipediaapi as api
import numpy as np
import matplotlib as plt
import wikipedia as wikip
import random as rd
```

```python
wiki = api.Wikipedia('Datenanalyse', 'de')
```

## Alle Einzelseiten der Kategorie Zweiter Weltkrieg

```python
memberliste = []
def print_categorymembers(categorymembers, namesp, level=0, max_level=10):
        for c in categorymembers.values():
            if c.ns == namesp and level < max_level:
                #print("%s: %s (ns: %d)" % ("*" * (level + 1), c.title, c.ns))
                memberliste.append(c.title)

kategorie = wiki.page("Kategorie:Zweiter Weltkrieg")
print_categorymembers(kategorie.categorymembers, 0)
print(len(memberliste))
```

## Alle Unterkategorien der Kategorie Zweiter Weltkrieg

```python
subcats = []
def print_subcategorymembers(categorymembers, namesp, level=0, max_level=10):
        for c in categorymembers.values():
            if c.ns != namesp and level < max_level:
                #print("%s: %s (ns: %d)" % ("*" * (level + 1), c.title, c.ns))
                subcats.append(c.title)

kategorie = wiki.page("Kategorie:Zweiter Weltkrieg")
print_subcategorymembers(kategorie.categorymembers, 0)
del subcats[0]
print(subcats)
```

## Alle Seiten der Unterkategorien und Kategorie Zweiter Weltkrieg

```python
pages_from_subc = []

for c in subcats:
    subcat = wiki.page(c)
    print_categorymembers(subcat.categorymembers, 0)

memberliste = list(set(memberliste))
print(len(memberliste))
```

## 100 zufällig gewählte Artikel der Kategorie Zweiter Weltkrieg und dessen Unterkategorien

```python
randomlist = rd.sample(memberliste,100)
randomlist
```

```python
backlinks = []

for p in randomlist:
    backlinks.extend(wiki.page(p).backlinks)
backlinks = list(set(backlinks))
backlinks
```

```python
korrekte_bl = []
for l in backlinks:
    categories = wiki.page(l).categories
    for c in categories:
        if c == 'Kategorie:Zweiter Weltkrieg':
            korrekte_bl.append(l)
korrekte_bl
```
