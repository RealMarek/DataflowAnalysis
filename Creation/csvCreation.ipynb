```python
import json

# Read the file content
with open('../Data/Köln/scrapedLinks.json', 'r') as file:
    file_content = file.read()

# Split the content into separate JSON-like objects
json_like_objects = file_content.split('}{')

# Add back the braces to each object to make them valid JSON
json_like_objects = ['{' + obj + '}' for obj in json_like_objects]

# Parse each object as JSON
for obj in json_like_objects:
    try:
        parsed_json = json.loads(obj)
        print(parsed_json)
    except json.JSONDecodeError as e:
        print(f"Error parsing JSON: {e}")
```

```python
json_like_objects[0] = json_like_objects[0][1:]
json_like_objects[len(json_like_objects)-1] = json_like_objects[len(json_like_objects)-1][:len(json_like_objects[len(json_like_objects)-1])-1]

failures = []
for i in range(len(json_like_objects)):
    try:
        parsed_json = json.loads(json_like_objects[i])
        print(parsed_json)
    except json.JSONDecodeError as e:
        print(f"Error parsing JSON: {e}")
        failures.append(i)

```

```python
objectList = [json.loads(e) for e in json_like_objects]
```

```python
pd_format = {"base":[],"link":[],"type":[]}
for e in objectList:
    for o in e["outgoing_links"]:
        pd_format["base"].append(e["baselink"])
        pd_format["link"].append(o)
        pd_format["type"].append("outgoing")
pd_format
```

```python
import pandas as pd
df = pd.DataFrame(pd_format)
df.to_csv("../Data/Köln/structure.csv",index=False)
```

```python
import urllib3
from bs4 import BeautifulSoup
import re  
import requests
import json
import ast
import numpy as np
import pandas as pd


def getOrderedLinks(wikipage):
    try:
        http = urllib3.PoolManager()
        response = http.request("GET", wikipage)
        soup = BeautifulSoup(response.data)
        soup = soup.find(id="mw-content-text")
        link_list = soup.find_all("a",href=re.compile("/wiki/"))
        return link_list
    except:
        print(wikipage)
    
    
def filterLinks(soup):
    reg = re.compile("^(?!/wiki/Datei:).+")
    reg2 = re.compile("^(?!/wiki/Spezial:).+")
    reg3 = re.compile("^/wiki")
    filteredList = [e["title"] for e in soup if reg.match(e["href"]) and reg2.match(e["href"]) and reg3.match(e["href"])]
    return filteredList

def filterTitle(titleList):
    return [e for e in titleList if not ":" in e]

def cutList(titleList, length):
    return list(set(titleList[:length]))

def saveLinks(base_link, outgoing_links, position, first):
    data = pd.DataFrame({
        "base": base_link,
        "link": outgoing_links,
        "type": "outgoing"
    })
    data.to_csv(
        position,
        mode="a",
        index=False,
        header=first
    )        
def createLinkFile(titelList,position):
    for i,titel in enumerate(titelList):
        ordered = cutList(filterLinks(getOrderedLinks("https://de.wikipedia.org/wiki/" + titel)), 30)
        if(i == 0):
            saveLinks(titel,ordered,position,True)
        else:
            saveLinks(titel,ordered,position,False)
```

```python
base = pd.read_csv("../Data/Köln/baseLinks.csv")
createLinkFile(base["links"],"importantStructureOutgoing.csv")
```
