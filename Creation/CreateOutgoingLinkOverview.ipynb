```python
import ast
import urllib3
from bs4 import BeautifulSoup
import re
import json
```

```python
with open("./Data/randomArticles.txt", "r", encoding="utf-8") as links:
    articles = links.read()
```

```python
def getOrderedLinks(wikipage):
    http = urllib3.PoolManager()
    response = http.request("GET", wikipage)
    soup = BeautifulSoup(response.data)
    soup = soup.find(id="mw-content-text")
    link_list = soup.find_all("a",href=re.compile("/wiki/"))
    return link_list
        
def cutList(titleList, length):
    return list(set(titleList[:length]))

def filterLinks(soup):
    reg = re.compile("^(?!:).+")
    reg2 = re.compile("^/wiki")
    filteredList = [e["title"] for e in soup if reg.match(e["href"]) and reg2.match(e["href"])]
    return filteredList
```
